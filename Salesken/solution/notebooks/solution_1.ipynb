# Problem 1
- Question : Write an algorithm or a group of algorithms to match the misspelt city name with the correct city name and return the list of unique ids for each misspelt city name.

# Algorithm
- Load the correct and misspelt city data in two different variable(using pandas dataframes).
- Iterate over the misspelt city data.
  - Given **misspelt city name and respective country**, get the **possible correct city list** from the correct city data using groupby method.
  - Use **Edit Distance** technique or **Sequence Matcher algorithm by Ratcliff and Obershelp**, i used second approach to find similarity score.
  - Find nearest match(city from correct list) by sorting depending on the edit distance score. 
  - Return the unique id of predicted city.

```python
import pandas as pd
from difflib import get_close_matches


def correct_city(misspelt_name, correct_city_list):
    return get_close_matches(misspelt_name, correct_city_list, n=1, cutoff=0.2)

def city_list_by_country(misspelt_data, correct_data):
    correct_data = pd.read_csv(correct_data)
    misspelt_data = pd.read_csv(misspelt_data)
    correct_group = correct_data.groupby("country")
    
    for item in misspelt_data.itertuples():
        correct_city_data = correct_data.iloc[correct_group.indices[item.country]]
        list_city = correct_city_data.name.to_list()
        match = correct_city(item.misspelt_name, list_city)
        #print(match)
        city_id = correct_city_data[correct_city_data.name == match[0]].id.to_list()[0]
        #print("Missplt word :", item.misspelt_name, "::", "Match :", match, ":: city", city_id)
        data = {"misspelt_name": item.misspelt_name, "correct_name":match[0],"country":item.country, "city_id":city_id}
        yield data

def city_ids(misspelt_data, correct_data):
    city_ids = []
    for corrected_dict in city_list_by_country(misspelt_data, correct_data):
        city_ids.append(corrected_dict['city_id'])
    return city_ids
```
